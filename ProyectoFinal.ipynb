## Funciones para preprocesado de imagenes

```python
# Funcion histograma
def histograma(ImG, MatrizRGB):
    MatrizRGB = ('b','g','r')
    for i,color in enumerate(MatrizRGB):
        hist = cv2.calcHist([ImG],[i],None,[256],[0,256])
        plt.plot(hist,color = color)
    plt.title('Histograma')
    plt.show()
# Filtrro 
def AclararImg(IM):
    kernel2 = np.ones((2,2))
    return  cv2.filter2D(IM,-1,kernel2)

def desc(img):
    sift = cv2.SIFT_create()
    keypoints_1, descriptors=sift.detectAndCompute(img,None) 
    return descriptors
```

## Red neuronal

```python
#%load_ext tensorboard

x_train= Fotos  
y_train= Etiquetas

x_test= Fotos2
y_test=Etiquetas2


def create_model():
    return tf.keras.models.Sequential([
        tf.keras.layers.Conv2D(30, (3,3), activation = 'relu', input_shape = (100,100,1)),
        tf.keras.layers.MaxPooling2D(2,2),
        tf.keras.layers.Conv2D(64, (3,3), activation = 'relu'),
        tf.keras.layers.MaxPooling2D(3,3),
        tf.keras.layers.Conv2D(128,(4,4), activation = 'relu'),
        tf.keras.layers.MaxPooling2D(3,3),
    ############Red Capa densa para clasificación############
        tf.keras.layers.Dropout(0.5),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(256, activation = 'relu'),
        tf.keras.layers.Dense(2, activation = 'softmax')
    ])

#def create_model():
#  return tf.keras.models.Sequential([
#    tf.keras.layers.Flatten(input_shape=(100, 100), name='layers_flatten'),     
#    tf.keras.layers.Dropout(0.5),
#    tf.keras.layers.Dense(512, activation='relu', name='layers_dense'),
#    tf.keras.layers.Dropout(0.2, name='layers_dropout'),
#    tf.keras.layers.Dense(2, activation='softmax', name='layers_dense_2')
#  ])

```

## Entrenamiento

```python
###Se compilan los medelos se se agrega la función de perdida

model = create_model()
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)

model.fit(x=x_train, 
          y=y_train, 
          epochs=200, 
          validation_data=(x_test, y_test), 
          callbacks=[tensorboard_callback])
model.save('Modelo_Poyecto_final.h5')
model.save_weights('Pesos.h5')
print("Modelo finalizado")
```

### Metrica de evaluación 

```python
width_shape = 100
height_shape = 100
batch_size = 32 
names = ['Terreno','Embalse']
```

```python
test_data_dir = 'C:/Users/Asus/Documents/VisionDeMaquina/ProyectoFinal/Validacion'  

test_datagen = ImageDataGenerator()

test_generator = test_datagen.flow_from_directory(
    test_data_dir,
    target_size=(width_shape, height_shape), 
    batch_size = batch_size,
    class_mode='categorical', 
    shuffle=False)


custom_Model= load_model("Modelo_Poyecto_final.h5")

predictions = custom_Model.predict_generator(generator=test_generator)

y_pred = np.argmax(predictions, axis=1)
y_real = test_generator.classes
```

```python
matc=confusion_matrix(y_real, y_pred)

plot_confusion_matrix(conf_mat=matc, figsize=(9,9), class_names = names, show_normed=False)
plt.tight_layout()
```
